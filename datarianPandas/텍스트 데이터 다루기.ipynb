```python
import pandas as pd
```

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
titanic.head()
```

```python
titanic['nameSplit'] = titanic['Name'].str.split(',')
titanic.head(2)
```

```python
titanic['surName'] = titanic['nameSplit'].str.get(0)
titanic.head(2)
```

```python
titanic[titanic['Name'].str.contains('Countess')]
```

```python
titanic['Name'].str.len()
```

```python
titanic['NameLength'] = titanic['Name'].str.len()
```

```python
titanic.loc[titanic['NameLength'].idxmax(),:]
```

```python
titanic['Sex'] = titanic['Sex'].replace({'male':'M', 'female':'F'})
```

### 정규표현식

```python
df = pd.DataFrame(['mouse', 'dog', 'house and parrot', '23'], columns=['values'])

df
```

```python
df['v1'] = df['values'].str.contains('dog')
df
```

```python
df['v2'] = df['values'].str.contains('og')
df
```

```python
df['v3'] = df['values'].str.contains('\d', regex=True)
df
```

```python
df2 = pd.DataFrame(['40', '40.0', '41', '35'], columns=['values'])
df2
```

```python
df2['v1-0'] = df2['values'].str.contains('.0')
df2['v1-1'] = df2['values'].str.contains('.0', regex=True)
df2['v1-2'] = df2['values'].str.contains('\.0', regex=True)
df2['v2'] = df2['values'].str.contains('.0', regex=False)
df2
```
