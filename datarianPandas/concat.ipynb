```python
import pandas as pd
```

```python
air_quality_no2 = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/air_quality_no2_long.csv")
air_quality_pm25 = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/air_quality_pm25_long.csv")
```

```python
air_quality_no2.head(2)
```

```python
air_quality_pm25.head(2)
```

```python
air_quality = pd.concat([air_quality_no2,air_quality_pm25], axis=0)
air_quality
```

```python
air_quality['parameter'].value_counts()
```

```python
print(air_quality.shape)
```

```python
# concat을 사용하고 데이터의 출처를 구분해야하는 경우
# 이런 상황을 방지하기 위해서 concat에 index를 넣어주는 기능이 있음
air_quality_index = pd.concat([air_quality_no2,air_quality_pm25], axis=0, keys=["PM2", "NO2"])
air_quality_index
```

```python
# 인덱스 값을 컬럼명으로 다루는 방법
air_quality_index.reset_index(level=0).rename(columns={"level_0":"air"})
```
