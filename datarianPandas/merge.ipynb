```python
import pandas as pd
```

```python
stations_coord = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/air_quality_stations.csv')
stations_coord
```

```python
parameter_coord = pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/air_quality_parameters.csv')
parameter_coord
```

```python
air_quality_no2 = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/air_quality_no2_long.csv")
air_quality_pm25 = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/air_quality_pm25_long.csv")

air_quality = pd.concat([air_quality_no2,air_quality_pm25], axis=0)
air_quality
```

```python
air_quality = pd.merge(air_quality, stations_coord, how='left', on='location')
air_quality.head(2)
```

```python
# 기준이 되는 컬럼의 이름이 다를 때

pd.merge(air_quality,parameter_coord, how='left', left_on='parameter', right_on='id')
```
