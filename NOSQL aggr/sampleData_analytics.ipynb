```python
# lookup 
results = db.transactions.aggregate([
    {'$lookup':{
        'from': 'accounts',
        'localField':'account_id',
        'foreignField' :'account_id',
        'as' : 'oo'
    }},
    {'$project':{'account_id':1, 'oo':1}},
    {'$limit': 3},
    {'$unwind': {'path' : '$oo'}},
    {'$unwind': {'path' : '$oo.products'}}
])

for result in results:
    pprint(result)
```
